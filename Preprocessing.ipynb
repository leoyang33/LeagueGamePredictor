```python
df_team = pd.read_csv('Desktop/LeagueGamePredictor-main/data/LCS 2023 Spring - Team Stats - OraclesElixir.csv')
df_team.head()
```

```python
df_team = df_team.drop(columns=['AGT', 'GP', 'K', 'D', 'KD', 'GPR', 'EGR', 'MLR', 'ELD%', 'FBN%', 'BN%', 'CWPM'])
df_team
```

```python
percents = ['GSPD', 'FB%', 'FT%', 'F3T%', 'HLD%', 'FD%', 'DRG%', 'LNE%', 'JNG%']

for stat in percents:
    df_team[stat] = list(map(lambda x: x[:-1], df_team[stat].values))
    df_team[stat] = [float(x) for x in df_team[stat].values]
    df_team[stat] = round(df_team[stat] / 100, 3)
    
df_team
```

```python
merged = df_team.join(df2.set_index('Team'), on='Team')
del merged['wcpm']
del merged['wpm']
del merged['kp']
del merged['kill_a']
del merged['death_a']
del merged['assist_a']
```
