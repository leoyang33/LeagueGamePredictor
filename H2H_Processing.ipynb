```python
import pandas as pd
import numpy as np
import itertools
```

```python
#all of the data is from 2023
df = pd.read_csv('Desktop/LeagueGamePredictor-main/data/2023_LoL_esports_match_data_from_OraclesElixir_20230421.csv')
```

```python
df = df.loc[df['league'].isin(['LCS', 'LEC', 'LCK'])]
df = df.loc[df['split'] == 'Spring']
df = df.loc[df['playoffs'] == 0]
df = df[['gameid', 'gamelength', 'result', 'kills', 'deaths', 'assists', 'doublekills', 'triplekills', 'quadrakills','pentakills','damagetochampions', 'dpm', 'damageshare',  'damagetakenperminute', 'wardsplaced', 'wpm', 'controlwardsbought', 'vspm','totalgold', 'earned gpm', 'earnedgoldshare', 'cspm', 'golddiffat10', 'xpdiffat10', 'csdiffat10', 'golddiffat15', 'xpdiffat15', 'csdiffat15' ]]
df = df.dropna(subset=['damageshare'])
#there are 381 games in this season
df.head()
#always in the order (top, jg, mid, bot, sup)
```

```python
unique_games = list(df['gameid'].unique())
count = len(unique_games)
```

```python
#first try to get code to work for one game
master_list = []

for i in range(count):
    first_game = df.loc[df['gameid'] == unique_games[i]]
    game_length = list(first_game['gamelength'])[0]
    #print(game_length)
    result = list(first_game['result'])[0]
    #print(result)
    gameid = list(first_game['gameid'])[0]
    #print(gameid)

    #result: 1 = blue team wins (first team), 0 = red team wins (second team)
    game = first_game.drop(['gameid','result', 'gamelength'], axis=1)

    Row_list =[]

    # Iterate over each row
    for i in range((game.shape[0])):

        # Using iloc to access the values of 
        # the current row denoted by "i"
        Row_list.append(list(game.iloc[i, :]))

    merged = list(itertools.chain(*Row_list))
    merged.append(game_length)
    merged.append(result)
    master_list.append(merged)

```

```python
#very last row is the result: 1 if blue won and 0 if red won
final = pd.DataFrame(master_list)
final.to_csv("Desktop/H2H_processing.csv")
```
