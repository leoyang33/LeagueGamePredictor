```python
#imports
import pandas as pd
import numpy as np
```

```python
#all of the data is from 2023
df = pd.read_csv('Desktop/LeagueGamePredictor-main/data/2023_LoL_esports_match_data_from_OraclesElixir_20230421.csv')
del df['url']
df.head()
```

```python
df2 = df.loc[df['league'].isin(['LCS', 'LEC', 'LCK','PCS','VCS','CBLOL','LJL','LLA'])]
df3 = df2.loc[df2['split'] == 'Spring']
df4 = df3.loc[df3['playoffs'] == 0]
df4
```

```python
#check for missing data
complete_check= df4['datacompleteness'].tolist()
#print(complete_check)
for x in complete_check:
    if x == 'partial':
        print('hello')
```

```python
df4.to_csv('Desktop/2023_max_regions_cleaned.csv')
```

```python
unique_games = list(df4['gameid'].drop_duplicates())
print(len(unique_games))
```

```python
df5 = df4.nlargest(1490, ['minionkills'])
df5.to_csv("Desktop/max_regions.csv")
```

```python
df5 = df5[['gameid','result', 'ckpm', 'gspd','golddiffat15', 'firstblood', 'firsttower', 'firsttothreetowers', 'turretplates', 'firstherald', 'firstdragon', 'dragons', 'opp_dragons', 'minionkills', 'monsterkills', 'wpm', 'wcpm', 'golddiffat10', 'cspm', 'csdiffat10', 'earned gpm']]
#peform the derivation - this will have to be done on the 200 as well
dragons = df5['dragons'].tolist()
opp_dragons = df5['opp_dragons'].tolist()
denom = [sum(x) for x in zip(dragons, opp_dragons)]
dragon_percent = [round(x/y,3) for x,y in zip(dragons, denom)]
df5['dragons'] = dragon_percent
df5 = df5.drop('opp_dragons', axis=1)
```

```python
agg = []
monster_agg = []
order_unique = []
for x in unique_games:
    temp = df5.loc[df5['gameid'] == x]
    values = temp['minionkills'].tolist()
    monster_values = temp['monsterkills'].tolist()
    total = 0
    total_monster = 0
    for i in values:
        total = total + i
    for y in monster_values:
        total_monster = total_monster + y
    agg.append(total)
    monster_agg.append(total_monster)
    order_unique.append(x)
    
        
```

```python
ordered = pd.DataFrame(order_unique, columns=['gameid'])
ordered['minion_percent'] = agg
ordered['monster_percent'] = monster_agg

natural_joined = df5.join(ordered.set_index('gameid'), on='gameid')
minion = list(natural_joined['minionkills'])
bottom = list(natural_joined['minion_percent'])
                             
                          
m_percent = [round(x/y,3) for x,y in zip(minion, bottom)]      
natural_joined['minionkills'] = m_percent
natural_joined = natural_joined.drop('minion_percent', axis=1)



monster = list(natural_joined['monsterkills'])
bottom_monster = list(natural_joined['monster_percent'])
                             
                          
monster_percent = [round(x/y,3) for x,y in zip(monster, bottom_monster)]      
natural_joined['monsterkills'] = monster_percent
natural_joined = natural_joined.drop('monster_percent', axis=1)
natural_joined = natural_joined.rename(columns={'golddiffat10': 'gd10', 'csdiffat10': 'csd10', 'earned gpm':'egpm'})
fix = natural_joined['egpm'].tolist()
fixed = []
for x in fix:
    fixed.append(round(x/5,3))
natural_joined['egpm'] = fixed
natural_joined
```

```python
natural_joined = natural_joined[natural_joined['gameid'] != 'ESPORTSTMNT03_3116248']
natural_joined = natural_joined[natural_joined['gameid'] != 'ESPORTSTMNT01_3332332']
natural_joined
```
